```python
dfs = [cnae, setor, vinculo,escolaridade,idade,total_pessoas]
df_final_ = dfs[0]

for df in dfs[1:]:
    df_final_ = pd.merge(df_final_, df, on=["capital", "faixa_remuneracao_media_sm", "sexo", "ano"], how='outer')
df_final_.head()
```

```python
df_final_.shape
```

```python
df_final_ = df_final_.drop(df_final_.filter(like='Ignorado').columns, axis=1)
df_final_ = df_final_.drop(df_final_.filter(like='Outros, ignorados').columns, axis=1)
```

```python
df_final_ = df_final_.fillna(0)
```

```python
df_final_.to_csv(f'{path}/df_final.csv', index=False,sep=',', decimal='.', float_format='%.2f')
```

```python
colunas_object = df_final_.select_dtypes(include=['object']).columns
# Supondo que estas sejam as colunas que você não quer alterar
colunas_a_manter = ['capital', 'faixa_remuneracao_media_sm', 'sexo']

# Atualizando a lista para excluir essas colunas
colunas_para_tratar = [col for col in colunas_object if col not in colunas_a_manter]

for coluna in colunas_para_tratar:
    df_final_[coluna] = pd.to_numeric(df_final_[coluna], errors='coerce')

```

```python
df_agrupado = df_final_.groupby(['capital', 'faixa_remuneracao_media_sm', 'sexo']).median().reset_index()
df_agrupado_rounded = df_agrupado.round(0).reset_index(drop=True)
df_agrupado_rounded.drop('ano', axis=1, inplace=True)
for coluna in df_agrupado_rounded.select_dtypes(include=['float', 'float64']).columns:
    df_agrupado_rounded[coluna] = df_agrupado_rounded[coluna].round(0).astype(int)

# Agora, mostrar as primeiras 10 linhas deve funcionar sem erros
df_agrupado_rounded.head(10)
```

```python
df_agrupado_rounded.to_csv(f'{path}/df_group_by_year.csv',index=False,sep=',', decimal='.')
```

```python
#criando  o df para fazer o feature selection com variaveis dummy 
df_dummies = pd.get_dummies(df_agrupado_rounded, columns=['capital'])
#df_dummies = pd.get_dummies(df_dummies, columns=['sexo'])
df_dummies['faixa_remuneracao_media_sm'] = df_dummies['faixa_remuneracao_media_sm'].apply(lambda x: int(x[:2]))

# Agora, vamos tratar os valores faltantes no dataframe. Para simplicidade, vamos preencher os valores faltantes com a média da coluna
#df_dummies.fillna(df_dummies.mean(), inplace=True)



df_dummies.head(20)
```
