```python
idade = pd.read_csv("idade.csv")
idade = idade.drop(idade.filter(like='{ñ class}').columns, axis=1)
idade.head()
```

```python
setor = pd.read_csv("IBGE_setor.csv")
#ok

setor.head()
```

```python
cnae = pd.read_csv("CNAE_grupo.csv")
cnae = cnae.drop(cnae.filter(like="CAÇA, REPOVOAMENTO CINEGÉTICO E SERVIÇOS RELACIONADOS").columns, axis=1)
#ok
cnae.head()
```

```python
dfs = [cnae, setor, vinculo,escolaridade,idade]
df_final_ = dfs[0]

for df in dfs[1:]:
    df_final_ = pd.merge(df_final_, df, on=["capital", "faixa_remuneracao_media_sm", "sexo", "ano"], how='outer')
df_final_.head()
```

```python
df_final_.to_csv(f'{path}/df_final.csv', index=False)
```

```python
#criando  o df para fazer o feature selection com variaveis dummy 
df_dummies = pd.get_dummies(df_final_, columns=['capital'])
#df_dummies = pd.get_dummies(df_dummies, columns=['sexo'])
df_dummies['faixa_remuneracao_media_sm'] = df_dummies['faixa_remuneracao_media_sm'].apply(lambda x: int(x[:2]))

# Agora, vamos tratar os valores faltantes no dataframe. Para simplicidade, vamos preencher os valores faltantes com a média da coluna
#df_dummies.fillna(df_dummies.mean(), inplace=True)



df_dummies.head(20)
```

```python
df_dummies.to_csv(f'{path}/df_dummies.csv', index=False)
```
