```python
def unstack_double_header(df, col_index):
    """
    transform df with double headers into single header
    """

    # Unstack two headers
    cols = df.iloc[col_index][1::].unique()
    cols = (list(set([col.strip() for col in cols])))

    result = pd.DataFrame()
    for col in cols:

        # get the indexes of the desired columns
        idx = np.where(df.iloc[0] == col)[0]
        # insert first column (key)
        idx = np.insert(idx, 0, 0, axis=0)

        # filter just the desired columns and removes lines from headers
        df_aux = df.loc[2::,idx]
        df_aux.columns = df.loc[1, idx]

        df_aux["new_col"] = col
        result = pd.concat([result, df_aux])

    return result
```

```python
df_final = pd.DataFrame()
for i in files_csv:
    
    df = pd.read_csv(i,sep=";",  encoding='latin-1',skiprows=1)
    ano = re.search(r'\d{4}', i).group()
    print(f"ingerindo arquivo {i} ano {ano} ")
    
    # Apagando tudo a partir de total
    indice_selecoes = df[df['Sexo Trabalhador'] == "Total"].index[0]
    df = df.drop(df.index[indice_selecoes:])

    #apagar todas as colunas com total no nome
    df = df.drop(df.filter(like='Total').columns, axis=1)
    df = df.drop(df.filter(like='{ñ class}').columns, axis=1)

    for coluna in df.columns:
        if df[coluna].iloc[0] == 'Total' or df[coluna].iloc[0] == '{ñ class}':
            # Excluindo a coluna se a primeira entrada for 'Total'
            df = df.drop(columns=[coluna])
    df.columns = new_cols
    
    df_sem_header = df.copy()
    df_sem_header.loc[-1] = df.columns
    df_sem_header.index = df_sem_header.index + 1
    df_sem_header = df_sem_header.sort_index()
    df = pd.DataFrame(df_sem_header.values)
    df.iloc[1,0]="sexo"

    linha_0 = df.iloc[0].copy()  # copia da linha 0
    df.iloc[0] = df.iloc[1]  # linha 0 recebe os valores da linha 1
    df.iloc[1] = linha_0  # linha 1 recebe os valores copiados da linha 0

    df= unstack_double_header(df,0)
    df = df.rename(columns={'new_col': 'faixa_salarial'})
    df['sexo'] = df['sexo'].replace({'01:Masculino': 'm', '02:Feminino': 'f'})
    df["ano"] = ano
    df_final = pd.concat([df_final, df], ignore_index=True)
    

    
df_final.to_csv('sexoXsetorXSM.csv', index=False)
```
