```python
#sera que nas regioes do brasil conseguimos enxergar uma dificuldade maior de se ter mulheres em faixas salarias especificas?

#df_regiao= df[["capital","faixa_remuneracao_media_sm","sexo","ano","total_pessoas_por_escolaridade"]]
df_regiao = pd.pivot_table(df, values='total_pessoas_por_escolaridade', index=['capital', 'faixa_remuneracao_media_sm',"ano"],
                       columns=['sexo'], aggfunc="sum")
df_regiao.reset_index(inplace=True)
df_regiao["proporcao_f"] = df_regiao["F"]/(df_regiao["F"]+df_regiao["M"])
df_regiao["proporcao_m"] = df_regiao["M"]/(df_regiao["F"]+df_regiao["M"])
df_regiao['entropia_F_em_relacao_a_M'] = (df_regiao['proporcao_f'] * np.log(1/df_regiao['proporcao_f'])) + (df_regiao['proporcao_m'] * np.log(1/df_regiao['proporcao_m']))

df_regiao_entropia= df_regiao
df_regiao_entropia.head()
```

```python
#Sera que em algum estado temos uma dificuldade maior de encontrar mulheres no mercado de trabalho?
#df_capital = df[["capital","sexo","ano","total_pessoas_por_escolaridade"]]
df_capital = pd.pivot_table(df, values='total_pessoas_por_escolaridade', index=['capital',"ano"],
                       columns=['sexo'], aggfunc="sum")
df_capital.reset_index(inplace=True)
df_capital["proporcao_f"] = df_capital["F"]/(df_capital["F"]+df_capital["M"])
df_capital["proporcao_m"] = df_capital["M"]/(df_capital["F"]+df_capital["M"])
df_capital['entropia_F_em_relacao_a_M'] = (df_capital['proporcao_f'] * np.log(1/df_capital['proporcao_f'])) + (df_capital['proporcao_m'] * np.log(1/df_capital['proporcao_m']))

#df_capital.head(10)
df_capital.loc[df_capital["capital"] == "São Paulo - SP"].sort_values(by=["ano"])

#nao tem uma entropia significativa para afirmar a desigualdade populacional trabalhadora por genero

#resultado = df_capital.groupby('capital')['entropia_F_em_relacao_a_M'].mean().reset_index()
#resultado
```

```python
#sera q as proporções por estado tem uma diferenca se olharmos media ou mediana?
df_capital.groupby('capital')['proporcao_f','proporcao_m'].median().reset_index()
#sim, temo alguns estados q tem uma diferenca percentual significativa "a olhos"
```

```python
#e se olharmos para os ultimos 3 anos de base?
#sera q as proporções por estado tem uma diferenca se olharmos media ou mediana?
df_capital_last_3y=df_capital[df_capital["ano"]>=2020]
df_capital_last_3y.groupby('capital')['proporcao_f','proporcao_m'].median().reset_index()
#sim, temo alguns estados q tem uma diferenca percentual, aqui o valor max de + homens é maior
```

```python
#qui-quadrado

from scipy.stats import chi2_contingency
```

```python
#vou usar primeiro a tabela com as faixas salariais 

df_faixa.head()
```

```python
tabela_contingencia = df_faixa[['proporcao_f', 'proporcao_m']].values
# Realize o teste qui-quadrado de independência.
chi2, p_valor, _, _ = chi2_contingency(tabela_contingencia)
# Imprima o valor de chi-quadrado e o p-valor.
print("Valor de chi-quadrado:", chi2)
print("P-valor:", p_valor)
```
