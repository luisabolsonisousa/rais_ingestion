```python
cnae = pd.read_csv("cnae.csv")

cnae = cnae.drop(cnae.filter(like="CAÇA, REPOVOAMENTO CINEGÉTICO E SERVIÇOS RELACIONADOS").columns, axis=1)
#ok
cnae.info()
```

```python
dfs = [cnae, setor, vinculo,escolaridade,idade,total_pessoas]
df_final_ = dfs[0]

for df in dfs[1:]:
    df_final_ = pd.merge(df_final_, df, on=["capital", "faixa_remuneracao_media_sm", "sexo", "ano"], how='outer')
df_final_.head()
```

```python
df_final_.shape
```

```python
df_final_ = df_final_.drop(df_final_.filter(like='Ignorado').columns, axis=1)
df_final_ = df_final_.drop(df_final_.filter(like='Outros, ignorados').columns, axis=1)
```

```python
df_final_ = df_final_.fillna(0)
```

```python
df_final_.to_csv(f'{path}/df_final.csv', index=False,sep=',', decimal='.', float_format='%.2f')
```

```python
df_final_[["capital",'faixa_remuneracao_media_sm', 'sexo',"total_pessoas","ano"]].head(20)
```

```python


df_agrupado = df_final_.groupby(['capital', 'faixa_remuneracao_media_sm', 'sexo']).median().reset_index()


# Se você quiser manter as colunas 'capital', 'faixa_remuneracao_media_sm', 'sexo' como colunas fixas e não incluí-las na mediana
# mas apenas usar 'ano' para agrupar, você pode fazer um ajuste, dependendo da sua necessidade exata.

df_agrupado.head()
#df_agrupado_rounded = df_agrupado.round(0).reset_index(drop=True)
#df_agrupado_rounded.drop('ano', axis=1, inplace=True)

```
